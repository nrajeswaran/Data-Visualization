### Deceptive Project Data Wrangling

```python
#Import Libraies
import pandas as pd

#Read file to Dataframe
Data = pd.read_csv("API_EN.ATM.CO2E.PC_DS2_en_csv_v2.csv",sep='\t')
```

```python
#Drop empty columns(axis=1 for columns and axis=0 for rows; 
#how='all' to drop if all values are nan and how='any' to drop if any value is nan)
Data = Data.dropna(axis=1,how='all')
```

```python
#Drop junk columns
columns_to_drop = ['Country Code', 'Indicator Name', 'Indicator Code']
Data = Data[[col for col in Data.columns if col not in columns_to_drop]]
```

```python
#To make a particular as the index of the table
Data = Data.set_index('Country Name')
```

```python
Data.head()
```

```python
#To re-organise the table structure as required
import datetime as dt
Data_Final = pd.DataFrame(columns=['Country Name','Year','CO2 emissions (metric tons per capita)'])
d =0
for i in range(len(Data)) : 
    ts = Data.index[i]
    for j in range(len(Data.iloc[i])): 
        year = (str(Data.columns.values[j]))
        Data_Final.loc[d]=( ts,  year , Data.iloc[i][j])
        d=d+1
```
