```python
import pandas as pd

#Read the whole workbook
Lab1_Doc = pd.ExcelFile("Lab1_Doc.xlsx")

#Read the sheet1
Sheet_2017 = pd.read_excel("Lab1_Doc.xlsx",sheetname=0)
#Read the sheet2
Sheet_2015 = pd.read_excel("Lab1_Doc.xlsx",sheetname=1)
#Read the sheet3
Sheet_2013 = pd.read_excel("Lab1_Doc.xlsx",sheetname=2)
```

```python
#Display the sheet names in the workbook
read_sheets_name = Lab1_Doc.sheet_names
read_sheets_name
```

```python
#Replace the header with the first row for Sheet 2015
#grab the first row for the header
new_header = Sheet_2015.iloc[0] 
#take the data less the header row
Sheet_2015 = Sheet_2015[1:] 
#set the header row as the Sheet2015 header
Sheet_2015.columns = new_header 
```

```python
#Replace the header with the first row for Sheet 2013
#grab the first row for the header
new_header = Sheet_2013.iloc[0] 
#take the data less the header row
Sheet_2013 = Sheet_2013[1:] 
#set the header row as the Sheet2013 header
Sheet_2013.columns = new_header 
```

```python
#Drop junk columns from the sheet2017
columns_to_drop1 = ['UNUSED' , 'UNUSED.1', 'Exclude', 'Unnamed: 13', '1st source link', '2nd source link', '3rd source', 'source name']
Sheet_2017 = Sheet_2017[[col for col in Sheet_2017.columns if col not in columns_to_drop1]]
```

```python
#Drop junk columns from the sheet2015
columns_to_drop2 = ['UNUSED' , 'Link to individual study' , 'Exclude' , '1st source link', '2nd source link', '3rd source', 'source name']
Sheet_2015 = Sheet_2015[[col for col in Sheet_2015.columns if col not in columns_to_drop2]]
```

```python
#Drop empty columns(axis=1 for columns and axis=0 for rows; 
#how='all' to drop if all values are nan and how='any' to drop if any value is nan)
Sheet_2015 = Sheet_2015.dropna(axis=1,how='all')
```

```python
#Drop junk columns from the sheet2013
columns_to_drop3 = ['UNUSED' , 'Link to individual study' , 'Exclude' , '1st source link', '2nd source link', '3rd source', 'source name']
Sheet_2013 = Sheet_2013[[col for col in Sheet_2013.columns if col not in columns_to_drop3]]

Sheet_2013 = Sheet_2013.dropna(axis=1,how='all')
```

```python
#Replace YEAR column values
#For Sheet_2017 0=2004, 8 = 2012, 9 = 2013, 10=2014, 11=2015, 12=2016, 13 = 2017, 14=latest
#For Sheet_2015 0=2004, 8 = 2012, 9 = 2013, 10=2014, 11=latest
#For Sheet_2013 0=2004, 8 = 2012, 9 = 2013
Sheet_2013.loc[Sheet_2013['YEAR'] == 0, 'YEAR'] = 2004
Sheet_2013.loc[Sheet_2013['YEAR'] == 1, 'YEAR'] = 2005
Sheet_2013.loc[Sheet_2013['YEAR'] == 2, 'YEAR'] = 2006
Sheet_2013.loc[Sheet_2013['YEAR'] == 3, 'YEAR'] = 2007
Sheet_2013.loc[Sheet_2013['YEAR'] == 4, 'YEAR'] = 2008
Sheet_2013.loc[Sheet_2013['YEAR'] == 5, 'YEAR'] = 2009
Sheet_2013.loc[Sheet_2013['YEAR'] == 6, 'YEAR'] = 2010
Sheet_2013.loc[Sheet_2013['YEAR'] == 7, 'YEAR'] = 2011
Sheet_2013.loc[Sheet_2013['YEAR'] == 8, 'YEAR'] = 2012
Sheet_2013.loc[Sheet_2013['YEAR'] == 9, 'YEAR'] = 2013
#Sheet_2015.loc[Sheet_2015['YEAR'] == 10, 'YEAR'] = 2014
#Sheet_2015.loc[Sheet_2015['YEAR'] == 11, 'YEAR'] = 'latest'
#Sheet_2015.loc[Sheet_2015['YEAR'] == 12, 'YEAR'] = 'latest'
#Sheet_2017.loc[Sheet_2017['YEAR'] == 13, 'YEAR'] = 2017
#Sheet_2017.loc[Sheet_2017['YEAR'] == 14, 'YEAR'] = 'latest'
```

```python
#Replace DATA SENSITIVITY Column Values
#For Sheet_2017 1. Just email address/Online information 20 SSN/Personal details 300 Credit card information 4000 Email password/Health records 50000 Full bank account details
#For Sheet_2015 1. Just email address/Online information 20 SSN/Personal details 300 Credit card information 4000 Email password/Health records 50000 Full bank account details
#For Sheet_2013 1. Just email address/Online information 2. SSN/Personal details 3. Credit card information 4.Email password/Health records 5. Full bank account details
Sheet_2013.loc[Sheet_2013['DATA SENSITIVITY'] == 1, 'DATA SENSITIVITY'] = 'Just email address/Online information'
Sheet_2013.loc[Sheet_2013['DATA SENSITIVITY'] == 2, 'DATA SENSITIVITY'] = 'SSN/Personal details'
Sheet_2013.loc[Sheet_2013['DATA SENSITIVITY'] == 3, 'DATA SENSITIVITY'] = 'Credit card information'
Sheet_2013.loc[Sheet_2013['DATA SENSITIVITY'] == 4, 'DATA SENSITIVITY'] = 'Email password/Health records'
Sheet_2013.loc[Sheet_2013['DATA SENSITIVITY'] == 5, 'DATA SENSITIVITY'] = 'Full bank account details'
```
